- also if you look up you ccan find the uniform distribution (critics) and normal distribution (audience)

```python
sns.set_style('darkgrid')
f, axes = plt.subplots(2, 2, figsize=(15, 15))

# Corrected kdeplot calls
k1 = sns.kdeplot(data=data, x='BudgetMillions', y='AudienceRating', ax=axes[0, 0])
k2 = sns.kdeplot(data=data, x='BudgetMillions', y='Criticrating', ax=axes[0, 1])

k1.set(xlim=(-20, 160))
k2.set(xlim=(-20, 160))

# Violin plot
z = sns.violinplot(data=data[data.Genre == 'Drama'], x='Year', y='Criticrating', ax=axes[1, 0])

# Corrected kdeplot with shade
k4 = sns.kdeplot(data=data, x='Criticrating', y='AudienceRating', fill=True, cmap='Reds', ax=axes[1, 1])
k4b = sns.kdeplot(data=data, x='Criticrating', y='AudienceRating', cmap='Reds', ax=axes[1, 1])

plt.show()
```

```python
# Set the style with custom background color
sns.set_style('dark', {'axes.facecolor': 'black'})

# Create a 2x2 subplot grid
f, axes = plt.subplots(2, 2, figsize=(15, 15))

# Plot [0,0]
k1 = sns.kdeplot(data=data, x='BudgetMillions', y='AudienceRating', 
                 fill=True, cmap='inferno', ax=axes[0, 0])
k1b = sns.kdeplot(data=data, x='BudgetMillions', y='AudienceRating', 
                  cmap='cool', ax=axes[0, 0])

# Plot [0,1]
k2 = sns.kdeplot(data=data, x='BudgetMillions', y='Criticrating', 
                 fill=True, cmap='inferno', ax=axes[0, 1])
k2b = sns.kdeplot(data=data, x='BudgetMillions', y='Criticrating', 
                  cmap='cool', ax=axes[0, 1])

# Plot [1,0] - Violin plot
z = sns.violinplot(data=data[data.Genre == 'Drama'], 
                   x='Year', y='Criticrating', ax=axes[1, 0])

# Plot [1,1]
k4 = sns.kdeplot(data=data, x='Criticrating', y='AudienceRating', 
                 fill=True, cmap='Blues_r', ax=axes[1, 1])
k4b = sns.kdeplot(data=data, x='Criticrating', y='AudienceRating', 
                  cmap='gist_gray_r', ax=axes[1, 1])

# Set limits
axes[0, 0].set(xlim=(-20, 160))
axes[0, 1].set(xlim=(-20, 160))

plt.show()
```
